```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30,40]) 
# x축의 데이터는 자동으로 [0, 1, 2, 3]으로 입력
plt.show()
# 그래프를 보여주기 위한 show 함수
```

```python
x = [0,1,2,3]
y = [10,20,30,40]
plt.plot(x, y,'red')
plt.show()
```

```python
import numpy as np
x = np.arange(0,2*np.pi, 0.01)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
# 2 개의 plot함수로 2개의 그래프 그리기
```

```python
plt.plot(x, np.sin(x),'--')
plt.plot(x, np.cos(x),':')
plt.show()
```

```python
plt.bar([1,2,3,4],[10,20,30,50])
plt.show()
```

```python
plt.bar(range(1,5), [10,20,30,40], width=0.3)
plt.show()
```

```python
plt.bar(range(1,8,2), [10,20,30,50], width=0.7)
plt.bar(range(2,9,2), [20,10,50,50], width=0.7)
plt.show()
```

```python
plt.barh(range(1,8,2), [10,20,30,50], height=0.7)
plt.barh(range(2,9,2), [20,10,50,50], height=0.7)
plt.show()
```

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
iris_df = pd.DataFrame(iris_dataset['data'], columns=['sepal length', 'sepal width','petal length', 'petal width'])
plt.scatter(iris_df['sepal length'], iris_df['sepal width'])
plt.show()
```

```python
iris_df['label'] = iris_dataset['target']
print(iris_df)
```

```python
setosa_df = iris_df[iris_dataset['target']==0]
plt.scatter(setosa_df['sepal length'], setosa_df['sepal width'])
plt.show()
```

```python
plt.scatter(iris_df['sepal length'],iris_df['sepal width'], c=iris_df['label'])
plt.colorbar()
plt.show()
```
